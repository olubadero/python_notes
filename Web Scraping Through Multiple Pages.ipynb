```python
from bs4 import BeautifulSoup as BS
```

```python
import requests
import pandas as pd
from pandas import Series, DataFrame
```

```python
url = 'https://books.toscrape.com/catalogue/page-1.html'
```

```python
response = requests.get(url)
response = response.content
```

```python
soup = BS(response, 'html.parser')
```

```python
soup.prettify()
```

```python
ol = soup.find('ol')
```

```python
article = ol.find_all('article', class_="product_pod")
```

```python
books = []

for data in article:
    image = data.find('img')
    title = image.attrs['alt']
    rating = data.find('p')
    rating = rating['class'][1]
    price = data.find('p', class_='price_color')
    price = float(price.text[1:]) 
    availability = data.find('p', class_='instock availability')
    availability = availability.text
    availability = availability.replace('\n','')
    print(availability)
    
    books.append([title, price, rating, availability])
    print(books)
```

```python
##Scraping more pages through looping to have an idea of the scraped data

books = []

for i in range(1, 50):
    url = f'https://books.toscrape.com/catalogue/page-{i}.html'
    response = requests.get(url)
    response = response.content
    soup = BS(response, 'html.parser')
    ol = soup.find('ol')
    article = ol.find_all('article', class_="product_pod")

    for data in article:
        image = data.find('img')
        title = image.attrs['alt']
        rating = data.find('p')
        rating = rating['class'][1]
        price = data.find('p', class_='price_color')
        price = float(price.text[1:]) 
        availability = data.find('p', class_='instock availability')
        availability = availability.text
        availability = availability.replace('\n','')
    
    books.append([title, price, rating, availability])
    
    
```

```python
##convert scraped data to Pandas DataFrame
Books = DataFrame(books, columns=['Title', 'Price', 'Availability', 'Rating'])
Books
```

```python
page = 1
formats_all = []
while page != 35:
      url = f"https://www.bookdepository.com/bestsellers?page={page}"
      response = requests.get(url)
      html = response.content
      soup = BS(html, "lxml")
      formats_on_page = soup.select("div.item-info p.format")
      for product_format in formats_on_page:
            formats_all.append(product_format.get_text())
      page = page + 1
formats_series = pd.Series(formats_all)
formats_series.value_counts()
```
